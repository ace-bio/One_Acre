### Background

Nigeria faces several challenges in crop production, particularly in maize farming. These challenges include:

- **Climate Variability**: Unpredictable weather patterns affect planting and harvesting times, leading to reduced yields.
- **Soil Degradation**: Poor soil health due to erosion, nutrient depletion, and lack of organic matter.
- **Pest and Disease Pressure**: High incidence of pests and diseases that affect crop health and productivity.
- **Limited Access to Inputs**: Farmers often lack access to quality seeds, fertilizers, and other essential inputs.
- **Infrastructure Deficiencies**: Poor infrastructure, including inadequate storage facilities and transportation networks, hampers efficient distribution and marketing of produce.

Addressing these challenges is crucial for improving maize yields and ensuring food security in Nigeria.

```python
# Group the dataframe by the 'hybrid' column and calculate the mean yield (yield_kg_ha) for each group
mean_hybrid_yield = df_ng.groupby('hybrid')['yield_kg_ha'].mean()

mean_hybrid_yield
```

```python
# alpha = 0.1

# Function to perform t-test and return p-value
def perform_ttest(df, column):
    df_vs_yield = df[[column, 'yield_kg_ha']].dropna(how='any')
    group1 = df_vs_yield[df_vs_yield[column] == df_vs_yield[column].unique()[0]]
    group2 = df_vs_yield.drop(group1.index)
    stat_score, p_value = ttest_ind(group1['yield_kg_ha'], group2['yield_kg_ha'], nan_policy='omit')
    return p_value.round(4)

# Calculate p-values for 'strat', 'hybrid', and 'striga'
p_value_strat = perform_ttest(df_ng, 'strat')
p_value_hybrid = perform_ttest(df_ng, 'hybrid')
p_value_striga = perform_ttest(df_ng, 'striga')

# Display p-values
print(f'independence within group p-value \n strat: {p_value_strat}, \n hybrid: {p_value_hybrid} \n striga:  {p_value_striga}')
```

Based on the p-values calculated (alpha of 0.10):


- For the feauture `hybrid`, the p-value is 0.1649. Since this p-value is greater than 0.10, we fail to reject the null hypothesis. This indicates that there is no statistically significant difference in yield between the groups defined by `hybrid`.

- For the variable `strat`, the p-value is 0.0063. Since this p-value is less than the significance level of 0.10, we reject the null hypothesis. This suggests that there is a statistically significant difference in yield between the groups defined by `strat`.



- For the variable `striga`, the p-value is 0.9601. Since this p-value is much greater than 0.10, we fail to reject the null hypothesis. This strongly suggests that there is no statistically significant difference in yield between the groups defined by `striga`.

# Building a Predictive Model for Crop Yield in Nigeria



Despite the limited number of samples from Nigeria, we will attempt to build predictive models to understand the key variables influencing crop yield outcomes. The primary objective of this exercise is to identify the most significant factors affecting crop yield in Nigeria.

Previously, we experimented with various models including Lasso and Linear Regression. However, these models demonstrated poor performance, with the best models achieving an R-squared value between 0.2 and 0.3. This underperformance is expected, given the weak linear relationships between the numeric features and the target variable, even after applying transformations such as logarithmic, square root, and cube root.

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Drop the target variable 'yield_kg_ha' from the dataframe to create feature set X
X = df_ng.drop(columns= ['yield_kg_ha', 'yield_kg_ha_sqrt', 'yield_kg_ha_cbrt', 'yield_kg_ha_log'], axis=True)

# Convert categorical variables into dummy/indicator variables
X_dummy = pd.get_dummies(X, drop_first=True)
```

#### Feature Engineering

In light of the limitations observed with linear models, the target variable 'yield_kg_ha' has been transformed into a categorical variable for the purposes of this study. Specifically, yields below 1800 kg/ha have been classified as 'low yield', while yields equal to or exceeding this threshold have been classified as 'high yield'.

```python

# Define a condition to classify yield as 'low' if it is less than 1800, otherwise 'high'
condition = df_ng['yield_kg_ha'] < 1800

# Create the target variable y based on the condition
y = np.where(condition, 'low', 'high')
```

### Splitting training and testing data

```python
# Split the data into training and testing sets with 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X_dummy, y, test_size = 0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the 'season_prec' column of the training data and transform it
X_train['season_prec'] = scaler.fit_transform(X_train[['season_prec']])

# Transform the 'season_prec' column of the test data using the already fitted scaler
X_test['season_prec'] = scaler.transform(X_test[['season_prec']])
```

### The model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Initialize and train the RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=11, random_state=1)
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)


# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Calculate accuracy, precision, recall, and F1 score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='high')
recall = recall_score(y_test, y_pred, pos_label='high')
f1 = f1_score(y_test, y_pred, pos_label='high')

# Display the results

print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

```python

# Plot the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['low', 'high'], yticklabels=['low', 'high'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

The model achieved the following metrics:

- **Accuracy**: 0.7391
- **Precision**: 0.7778
- **Recall**: 0.6364
- **F1 Score**: 0.7

The confusion matrix is displayed above.
